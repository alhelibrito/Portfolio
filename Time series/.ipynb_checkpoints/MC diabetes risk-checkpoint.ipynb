## Evaluating Choice of Distribution

Prior to the Monte Carlo simulation of the logistic regression model, the empirical distributions of selected continuous predictor variables were examined. Several candidate probability distributions (Laplace, Normal, Exponential, Log-normal, Gamma, and Beta) were fitted to each predictor, and goodness of fit was assessed using the Kolmogorov–Smirnov (KS) test.

The KS statistic and corresponding p-value were computed for each fitted distribution and variable. This evaluation was conducted for comparative purposes, with the goal of informing the stochastic generation of predictors in the simulation framework rather than performing formal distributional in



```python
import scipy.stats as st

# List of predictor variables to evaluate
variables = ['imc','colest','trig','glu_suero','insulina']

# Candidate distributions to test
list_of_dists = ["laplace", "norm", "expon", "lognorm", "gamma", "beta"]

# Dictionary to store results
results = {}

for var in variables:
    print(f"\nEvaluating distributions for: {var}")
    data = dia[var]   # data for the current variable
    
    results[var] = []
    
    for dist_name in list_of_dists:
        # Dynamically get the distribution object from scipy.stats
        dist = getattr(st, dist_name)
        
        # Fit the distribution to the data (estimate parameters)
        params = dist.fit(data)
        
        # Perform the Kolmogorov-Smirnov test comparing data vs fitted distribution
        ks_result = st.kstest(data, dist_name, args=params)
        
        # Save results (distribution name, KS statistic, p-value)
        results[var].append((dist_name, ks_result.statistic, ks_result.pvalue))
        
        # Print results in a readable format
        print(f"  {dist_name}: KS={ks_result.statistic:.3f}, p={ks_result.pvalue:.3f}")

```

## Distribution Fit Summary

Based on the Kolmogorov–Smirnov test:

- **imc, colest, trig** → Log-normal distribution provided the best fit.  
- **glu_suero, insulina** → No adequate parametric fit was found.

**Conclusion:**  
Log-normal distributions are used for *imc*, *colest*, and *trig* in the Monte Carlo simulation.  
Bootstrap resampling is applied for *glu_suero* and *insulina* to avoid distributional assumptions.
